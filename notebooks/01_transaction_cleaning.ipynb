```python
#Step 2 — Create the “core transactions” dataframe
```

```python
import pandas as pd

df_tx = df.rename(columns={
    "Transaction_ID": "transaction_id",
    "Account_Number": "account_id",
    "Transaction_Date": "date",
    "Transaction_Amount": "amount",
    "Merchant_Name": "merchant",
    "Category": "raw_category"
})

df_tx = df_tx[["transaction_id", "account_id", "date", "amount", "merchant", "raw_category"]].copy()

df_tx.head(3)
```

```python
#Step 3 — Parse dates
```

```python
df_tx["date"] = pd.to_datetime(df_tx["date"], errors="coerce")

invalid_dates = df_tx["date"].isna().sum()
invalid_dates
```

```python
df_tx["date"].min(), df_tx["date"].max()
```

```python
#Step 4 — Standardize amounts 
```

```python
df_tx["amount"] = pd.to_numeric(df_tx["amount"], errors="coerce")

missing_amounts = df_tx["amount"].isna().sum()
missing_amounts
```

```python
df_tx["amount"].describe()
```

```python
#Step 5 — Clean text fields (merchant/category)
```

```python
df_tx["merchant"] = df_tx["merchant"].astype(str).str.strip().str.lower()
df_tx["raw_category"] = df_tx["raw_category"].astype(str).str.strip().str.lower()
```

```python
(df_tx["merchant"].isin(["", "nan", "none"])).sum()
```

```python
#Step 6 — Remove duplicates
```

```python
dup_count = df_tx.duplicated(subset=["date", "amount", "merchant"]).sum()
dup_count
```

```python
before = len(df_tx)
df_tx = df_tx.drop_duplicates(subset=["date", "amount", "merchant"]).copy()
after = len(df_tx)

before, after, before - after
```

```python
#Step 7 — Missing value report
```

```python
df_tx.isna().sum()
```

```python
#Step 8 — Save cleaned transactions output
```

```python
output_path = "data/cleaned/transactions_clean.csv"
df_tx.to_csv(output_path, index=False)

output_path
```

```python
import pandas as pd

df_tx = pd.read_csv("data/cleaned/transactions_clean.csv")
accounts_df = pd.read_csv("data/cleaned/accounts_clean.csv")

df_tx.shape, accounts_df.shape
```

```python
account_ids = accounts_df["account_id"].astype(str).unique().tolist()
selected_accounts = account_ids[:5] if len(account_ids) >= 5 else account_ids

templates = [
    {"current_balance": 5200.00,  "interest_rate": 22.9, "minimum_payment": 160.00},
    {"current_balance": 12400.00, "interest_rate": 7.2,  "minimum_payment": 275.00},
    {"current_balance": 8600.00,  "interest_rate": 9.9,  "minimum_payment": 210.00},
    {"current_balance": 3100.00,  "interest_rate": 19.5, "minimum_payment": 95.00},
    {"current_balance": 17800.00, "interest_rate": 5.9,  "minimum_payment": 320.00},
]

debts_rows = []
for i, acc in enumerate(selected_accounts):
    t = templates[i % len(templates)]
    debts_rows.append({
        "debt_id": f"D{i+1}",
        "account_id": str(acc),
        "current_balance": float(t["current_balance"]),
        "interest_rate": float(t["interest_rate"]),
        "minimum_payment": float(t["minimum_payment"]),
    })

df_debts = pd.DataFrame(debts_rows)
df_debts
```

```python
output_path = "data/cleaned/debts_clean.csv"
df_debts.to_csv(output_path, index=False)

import os
os.path.exists(output_path), len(df_debts)
```
